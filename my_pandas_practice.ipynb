```python
import pandas as pd
import numpy as np
print("#產生DataFrame")
df = pd.DataFrame(data=np.array([(7,8,9),(1,2,3),(4,5,6)]),columns=["A","B","C"])
print("列出df檢視")
print(df)

print("新增欄位")
df["D"]=[1,2,3]
print(df)

print("重新命名欄位")
df.columns=["1","2","3","4"]
print(df)


print("取出欄位名稱")
print(df.columns)
print("取出陣列內容值")
print(df.values)
print("取出陣列索引")
print(df.index)
print("列出陣列維度")
print(df.shape)
print("資料轉置，索引與欄位翻轉")
print(df.T)


print("雙索引範例")
tuples = [
    ("cobra","mark i"),("cobra","mark ii"),
    ("sidewinder","mark i"),("sidewinder","mark ii"),
    ("viper","mark ii"),("viper","mark iii")
]
print(tuples)
index = pd.MultiIndex.from_tuples(tuples)
values = [[12,2],[0,4],[10,20],
    [1,4],[7,1],[16,36]]
df = pd.DataFrame(values,columns = ["max_spped","shield"],index=index)
print(df)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[4,9]]*3,columns=["A","B"])
print(df)

print("-----------------------")
print("df和apply的應用")
print(df.apply(np.sqrt))

print("-----------------------")
print("column 做 sum")
print(df.apply(np.sum,axis=0))

print("-----------------------")
print("row 做 sum")
print(df.apply(np.sum,axis=1))

print("-----------------------")
print("沿著row 用lambda變數，結果像Series")
print(df.apply(lambda x:[1,2],axis=1))

print("-----------------------")
print("沿著row 用lambda變數，expand使list結果延伸為dataframe的column")
print(df.apply(lambda x:[1,2],axis=1,result_type="expand"))

print("-----------------------")
print("沿著row 用lambda變數，類似result_type=\"expand\"")
print("但column name會是series的index")
print(df.apply(lambda x:pd.Series([1,2],index=["foo","bar"]),axis=1))

print("-----------------------")
print("沿著row 用lambda變數，類似result_type=\"broadcast\"")
print("XXXXX")
print(df.apply(lambda x:[1,2],axis=1,result_type="broadcast"))
```
